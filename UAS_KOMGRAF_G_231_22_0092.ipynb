**A.) Distribusi data pada setiap atribut Purchase Price, Current Price dan Gain/Loss dengan grafik histrogram**



```python
# Membuat histogram untuk Gain/Loss
plt.figure(figsize=(10, 5))
plt.hist(dt['Purchase Price ($)'], bins=13, color='blue', alpha=1)
plt.title('Distribusi Gain/Loss')
plt.xlabel('Gain/Loss')
plt.ylabel('Frekuensi')
plt.show()

# Membuat histogram untuk Current Price
plt.figure(figsize=(10, 5))
plt.hist(dt['Current Price ($)'], bins=10, color='purple', alpha=1)
plt.title('Distribusi Current Price')
plt.xlabel('Current Price')
plt.ylabel('Frekuensi')
plt.show()

# Membuat histogram untuk Gain/Loss
plt.figure(figsize=(10, 5))
plt.hist(dt['Gain/Loss ($)'], bins=13, color='orange', alpha=1)
plt.title('Distribusi Gain/Loss')
plt.xlabel('Gain/Loss')
plt.ylabel('Frekuensi')
plt.show()
```

**B. Atribut Stock Name dan Symbol yang memiliki Dividen / Share 5 tertinggi**



```python
# Mengambil data
sorted_dt = dt.sort_values('Dividend/share', ascending=False)

# Memfilter data, dan memberi nomor urut
top5ds = sorted_dt.head(5)
top5ds = top5ds.set_index(pd.Index(range(1, len(top5ds)+1)))

# Memberi nama
hasil = top5ds[['Stock Name', 'Symbol','Dividend/share']]

print(hasil)

```

**C. Atribut Stock Name dan Symbol yang memiliki Market Value 5 tertinggi.**



```python
# Mengambil data
sorted_dt = dt.sort_values('Market Value ($)',ascending=False).reset_index(drop=True)

# Memfilter data, dan memberi nomor urut
top5mv = sorted_dt.head(5)
top5mv = top5mv.set_index(pd.Index(range(1, len(top5mv)+1)))

# Memberi nama
hasil = top5mv[['Stock Name', 'Symbol','Market Value ($)']]

print(hasil)
```

**D. Atribut Stock Name dan Symbol yang memiliki Annual Yield 7 tertinggi.**



```python
# Mengambil data
sorted_dt = dt.sort_values('Annual Yield (%)',ascending=False)

# Memfilter data, dan memberi nomor urut
top7ay = sorted_dt.head(7)
top7ay = top7ay.set_index(pd.Index(range(1, len(top7ay)+1)))

# Memberi nama
hasil = top7ay[['Stock Name', 'Symbol','Annual Yield (%)']]

print(hasil)
```
